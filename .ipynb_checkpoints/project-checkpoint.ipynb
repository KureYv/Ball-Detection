```python
import cv2 as cv
import numpy as np
```

```python
def process(feed):
    hsv = cv.cvtColor(feed, cv.COLOR_BGR2HSV)
    mask = cv.inRange(hsv, np.array([5, 150, 150]), np.array([20, 255, 255]))
    cannyEdge = cv.Canny(mask, 50, 150)
    return cannyEdge
```

```python
def least_squares_fit(contours):
    x = contours[:, 0]
    y = contours[:, 1]
    A = np.column_stack([2*x, 2*y, np.ones_like(x)])
    b = x**2 + y**2
    return find_unknowns(A, b)
```

```python
def find_unknowns(A, b):
    result, _, _, _ = np.linalg.lstsq(A, b)
    center_x, center_y = result[0], result[1]
    radius = np.sqrt(result[2] + center_x**2 + center_y**2)
    return center_x, center_y, radius
```

```python
def findContour(edges):
    contours, _ = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
    for contour in contours:
        center_x, center_y, radius = least_squares_fit(contour[:,0,:])
        if (radius < 10 or radius > 150):
            continue
        return center_x, center_y, radius
    return None, None, None
```

```python
cam = cv.VideoCapture(0)

while True:
    ret, frame = cam.read()
    edges = process(frame)
    center_x, center_y, radius = findContour(edges)
    if center_x is None:
        continue
    cv.rectangle(frame, (int(center_x - radius), int(center_y - radius)), (int(center_x + radius), int(center_y + radius)), (0, 255, 0), 2)
    cv.imshow('camera', frame)
    if cv.waitKey(1) == ord('q'):
         break
cam.release()
cv.destroyAllWindows()
```
